```python
import numpy as np
import cv2
import os
import scipy
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage

original_image = cv2.imread('batman_logo.jpg',0)

canny_edges = cv2.Canny(original_image, 100, 120)

# im = scipy.misc.imread('bike.jpg')
# im = im.astype('int32')
# dx = ndimage.sobel(img, 0)  # horizontal derivative
# dy = ndimage.sobel(img, 1)  # vertical derivative
# mag = np.hypot(dx, dy)  # magnitude
# mag *= 255.0 / np.max(mag)  # normalize (Q&D)
# scipy.misc.imsave('sobel.jpg', mag)

# mag = cv2.imread('sobel.jpg',0)

sobel_edges = ndimage.sobel(original_image, mode = 'reflect', cval = 0.9)

fig = plt.figure(figsize = (20, 15))

plt.subplot(221),plt.imshow(original_image, cmap = 'gray')
plt.title('Original Image in grayscale'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(original_image,cmap = 'Reds')
plt.title('Original Image in Red colors'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(canny_edges,cmap = 'Reds')
plt.title('Canny Edge Detector in Red colors'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(sobel_edges,cmap = 'Reds')
plt.title('Sobel Edge Detector in Red colors'), plt.xticks([]), plt.yticks([])

plt.show()
```
