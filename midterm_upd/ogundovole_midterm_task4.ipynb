```python
import numpy as np
import cv2
import os
import scipy
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage

original_image = cv2.imread('batman_logo.jpg',0)

canny_edges = cv2.Canny(original_image, 100, 101)

sobel_edges = ndimage.sobel(original_image, mode = 'reflect', cval = 0.9)

fig = plt.figure(figsize = (20, 15))

plt.subplot(221),plt.imshow(original_image, cmap = 'gray')
plt.title('Original Image in grayscale'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(original_image,cmap = 'Reds')
plt.title('Original Image in Red colors'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(canny_edges,cmap = 'Reds')
plt.title('Canny Edge Detector in Red colors'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(sobel_edges,cmap = 'Reds')
plt.title('Sobel Edge Detector in Red colors'), plt.xticks([]), plt.yticks([])

plt.show()

# commentary on results:
# Canny edges detection works similarly to a Band Pass Filter
# greater difference between upper & lower values gives LESS detail
# smaller difference gives MORE details, meaning sometimes (in this example)
# more pixels that are not necessary to represent the edges of the Batman Logo
# Sobel edge detector doesn't give much variables to adjust it's results
# this method produces a lot of pixels (looks very noisy), but the 'mode' and 'cval'
# parameter adjustments didn't give much difference
```
