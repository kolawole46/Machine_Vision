```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimage

image = pltimage.imread('.\\6_26.jpg')

fig = plt.figure(figsize=(20,15))

plt.subplot(221)
plt.imshow(image[200:600,810:1235])

plt.subplot(222)
plt.imshow(image)

theHead = image[200:600,810:1235]
compositeImage = image

for i in range(200,600):
    for j in range(1235,1660):
        for k in range(0,3):
            compositeImage[i,j,k] = theHead[i-200,j-1235,k]

plt.subplot(223)
plt.imshow(compositeImage)
plt.show()

# add cat' head somewhere else on the image
```
